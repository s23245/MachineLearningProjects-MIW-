```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = np.loadtxt('Sharp_char.txt')

x = data[:, [1]]
y = data[:, [0]]
```

```python
theta = np.hstack([x, np.ones(x.shape)])
theta2 = np.hstack([x*x, x, np.ones(x.shape)])
theta3 = np.hstack([1/x, np.ones(x.shape)])

V = np.linalg.pinv(theta) @ y
V2 = np.linalg.pinv(theta2) @ y
V3 = np.linalg.pinv(theta3) @ y

e = y - (V[0]*x + V[1])
sse = e.T@e
print(f'Linear regression = {round(sse[0][0], 2)}')

e2 = y - (V2[0] * x ** 2 + V2[1]*x + V2[2])
sse2 = e2.T@e2
print(f'Quadratic regression = {round(sse2[0][0], 2)}')

e3 = y - (V3[0] * (1/x) + V3[1])
sse3 = e3.T@e3
print(f'Hyberbolic regression = {round(sse3[0][0], 2)}')

plt.plot(x, y, 'r*')
plt.plot(x, V[0] * x + V[1])
plt.plot(x, V2[0] * x ** 2 + V2[1]*x + V2[2])
plt.plot(x, V3[0] * (1/x) + V3[1])
```
